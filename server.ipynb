```python
from flask import Flask, jsonify, request
import import_ipynb
import db
import api

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello, World!'

@app.route('/accessibility')
def accessibility():
  return jsonify({'status': 'good'})

from flask import abort

@app.route('/users/<email>', methods=['GET', 'POST', 'DELETE'])
def user(email):
  if request.method == 'GET':
    try:
      u = db.get_user(email)
      return jsonify(u)
    except Exception as e:
      return jsonify(e)

  if request.method == 'POST':
    try:
      content = request.get_json()
      password = content['password']
      db.create_user(email, password)
      return jsonify({'success': True})
    except Exception as e:
      return jsonify(e)

  if request.method == 'DELETE':
    try:
      db.delete_user(email)
      return jsonify({'success': True})
    except Exception as e:
      return jsonify(e)
  
@app.route('/users/<email>/certificates', methods=['GET', 'POST'])
def user_certificate(email):
  if request.method == 'GET':
    try:
      c = db.get_certificates_by_email(email)
      return jsonify(c)
    except Exception as e:
      return jsonify(e)
    
    return jsonify({'success': True})
  if request.method == 'POST':
    try:
      content = request.get_json()
      private_key = content['private_key']
      certificate = content['certificate']
      db.create_certificate(email, private_key, certificate)
      return jsonify({'success': True})
    except Exception as e:
      return jsonify(e)    

@app.route('/certificates/<certificate_id>', methods=['PATCH'])
def certificate(certificate_id):
  if request.method == 'PATCH':
    try:
      content = request.get_json()
      isActive = content['active']
      db.update_certificate_status(certificate_id, isActive)
      c = db.get_certificate_by_id(certificate_id)
      # notify an external system
#       make_post('http://httpbin.org/post', jsonify(c))
      return jsonify(c)
    except Exception as e:
      return jsonify(e)
    
app.run(host='0.0.0.0', port=8080, debug=False)
```
